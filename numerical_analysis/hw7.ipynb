```python


def plotBL(rho, sigma, axisbox = [-2, 2, -2, 2]):
    theta = linspace(0, 2*pi, 1000)
    eitheta = exp(1j * theta)
    z = rho(eitheta) / sigma(eitheta)
    plot(z.real, z.imag, 'r', linewidth=2)
    
    # plot axes
    xa, xb, ya, yb = axisbox
    plot([xa, xb], [0, 0], 'k')
    plot([0, 0], [ya, yb], 'k')
    axis(axisbox)
    axis('scaled')  # so circles are circular


```

```python
#1 step adam bashforth 

rho = lambda z: (z-1.)
sigma = lambda z: 1.
plotBL(rho,sigma)

```

```python
#2 step Adam-bashforth
rho = lambda z: z*z- z
sigma = lambda z: -0.5+1.5*z
plotBL(rho,sigma)
```

```python
#3-step adam-bashforth
rho = lambda z: (z-1.) * z**2
sigma = lambda z: ((23*z - 16)*z + 5.) / 12.
plotBL(rho,sigma)
```

```python
#1-step adam-moulton
rho = lambda z: z-1.
sigma = lambda z: 0.5+0.5*z
plotBL(rho,sigma)
```

```python
#2-step adam-moulton
rho = lambda z:  z**2-z
sigma = lambda z: ((5*z + 8)*z - 1.) / 12.
plotBL(rho,sigma)
```

```python
#3-step adam-moulton
rho = lambda z: (z-1.) * z**2
sigma = lambda z: ((9*z*z + 19*z - 5)*z + 1.) / 24.
plotBL(rho,sigma)
```
