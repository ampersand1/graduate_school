```python
import numpy as np                
import matplotlib.pyplot as plt
from numpy import linalg as LA
import scipy.linalg


# Values to set.  
n=int((1.0/0.01)-1)
h=0.01
tol = 0.00001 
kmax= 1000


A = np.eye(n*n)*-4+np.eye(n*n,n*n,k=n) + np.eye(n*n,n*n,k=-n)+np.eye(n*n,n*n,k=1)+np.eye(n*n,n*n,k=-1)
#print(A) #for testing.

for i in range(len(A)):
    if i > 0 and i% n == 0:
        A[i-1][i]=0
        A[i][i-1]=0       
#print(A) #for testing

u=np.zeros((n*n,1)) 
#print(u)

f = np.zeros((n*n,1))

i=1
while i < n*n+1:
    for j in range(1,n+1):
        f[i-1][0] = -2.0*np.pi*np.pi*np.sin(i*h*np.pi)*np.sin(j*h*np.pi)
        i=i+1
    
Utrue = np.zeros((n*n,1))        
i=1
while i < n*n+1:
    for j in range(1,n+1):
        Utrue[i-1][0] = np.sin(i*h*np.pi)*np.sin(j*h*np.pi)
        i=i+1
    
    
Permu, Lower, Upper = scipy.linalg.lu(A)

def ichol(a,b):
    n= len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != 0 and b[i][j] !=0:
                a[i][j] = a[i][j]
            if a[i][j] == 0 or b[i][j] ==0:
                a[i][j] = 0  
    return a

#print(Upper) 

C=ichol(Upper, A)        
M=np.dot(C.transpose(),C)    


rold = np.subtract(f,np.dot(A,u))
Zold = np.linalg.solve(M, rold)
    
p=Zold

file = open('/Users/jasmin/Documents/cmse841/hw6/pcg2.dat', 'r+')
error1 = np.amax(abs(Utrue))   
file.write('0' + '\t' + repr(error1)+ '\n')


#print(f)
#np.dot(A,u)

for k in range(1,kmax):
    w= np.dot(A,p)
    alpha1 = np.dot(Zold.transpose(),rold)
    alpha2 = np.dot(p.transpose(),w)
    alpha = alpha1/alpha2
    u= np.add(u, alpha*p)
    r= np.subtract(rold, alpha*w)
    if LA.norm(r) < tol : 
        break
    Z= np.linalg.solve(M, r)
    beta= np.dot(Z.transpose(),r)/np.dot(Zold.transpose(),rold)
    p=np.add(Z,beta*p)
    rold=r
    Zold=Z
    error = np.amax(abs(np.subtract(Utrue,u)))/error1
    file.write(repr(k)+ '\t'+ repr(error)+ '\n')
```
